## Step 1 prepare your data

Prepare the training data in the format of ```data_training_example.csv```, where each row represents a record. The content to be recorded includes internal variables (adjustable parts of the detector), external variables (characteristics of the external light source), and the detector response. The same format is required for the test data, which is ```data_testing_example.csv```.

````python
import numpy as np
import pandas as pd
data_training = pd.read_csv('data_training_example.csv')
data_testing = pd.read_csv('data_testing_example.csv')
````

````python
data_training.head()
````

## Step 2 Instantiating the SpectralReconstruction class

This class ```SpectrumReconstructionBasic``` is defined to implement the basic functions of spectral reconstruction. When instantiated, there are parameters  ```training_data```, ```internal_var_col_name```, ```external_var_col_name```, ```dependent_var_col_name```, and ```base_func``` to be passed in. The parameter ```verify_pivot_data``` controls the integrity check of the data, and the data check is performed as needed.

````python
from SpectrumReconstruction import SpectrumReconstructionBasic as SRBasic

sr = SRBasic(
    data_training,
    "V_bias/mV",
    "Temperature",
    "Current",
    "blackbody",
    verify_pivot_data=True
)
````

## Step 3 Pass in the test data and reconstruct the spectrum

The ```reconstruct_spectrum``` method of the ```SpectrumReconstructionBasic``` class is used to pass in the ```testing_data```, ```method```, and some necessary parameters to perform spectral reconstruction calculations.



````python
sr.reconstruct_spectrum(
    data_testing,
    'l1',
    lambda_reg=0.001,
    pass_in_pivot_data=False
)
````

## Step 4 Visualization of the reconstructed spectrum (optional)

Using ```plotly``` to visualize the reconstructed spectrum. The x-axis is the wavelength, and the y-axis is the spectrum intensity. The wavelength range is set from 10 nm to 20000 nm, and the spectrum intensity is calculated using the ```spectrum``` method of the ```SpectrumReconstructionBasic``` class.

````python
import plotly.express as px

wavelength = np.array(range(10, 20001, 10)) * 1e-9
spectrum = sr.spectrum(wavelength, True)

fig = px.line(x=wavelength, y=spectrum)
fig.show()
````
